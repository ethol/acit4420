## Maintaining you dataframe
INSERT UPDATE DELETE

```python
# insert 

d = {
    'survived': [0],
    'pclass': [3],
    'sex': ["male"],
    'age': [42.0],
    'sibsp': [0],
    'parch': [2],
    'fare': [3.50],
    'embarked': ["S"],
    'class': ["Third"],
    'who': ["man"],
    'adult_male': [True],
    'deck': ["C"],
    'embark_town': ["Oslo"],
    'alive': ["yes"],
    'alone': [True]
    
}
df = pd.DataFrame(data=d)
df
titanic_extended_edition = pd.concat([titanic, df])
titanic_extended_edition = titanic_extended_edition.reset_index(drop=True)
titanic_extended_edition
```

## Best practice: Make sure your index is unique! in the dataframe 

```python
#Update
# Cell
titanic_extended_edition.at[891, "survived"] = 1
titanic_extended_edition

# #Multicell
# titanic_extended_edition.loc[891,['survived', 'pclass']] = [1,3]
# titanic_extended_edition.loc[890:891,['survived', 'pclass']] = [1,3]

titanic_extended_edition
#row
actual_guy =  pd.Series({'survived': 0,
    'pclass': 3,
    'sex': "male",
    'age': 42.0,
    'sibsp': 0,
    'parch': 2,
    'fare': 3.50,
    'embarked': "S",
    'class': "Third",
    'who': "man",
    'adult_male': True,
    'deck': "C",
    'embark_town': "Oslo",
    'alive': 1,
    'alone': True
})

titanic_extended_edition.iloc[1] = actual_guy
titanic_extended_edition
```

```python
# Delete
#Row
titanic_extended_edition.drop(0)
#Column
titanic_extended_edition.drop("pclass", axis=1)
```
